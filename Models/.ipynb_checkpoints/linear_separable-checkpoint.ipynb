```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
iris.items()
```

```python
X = iris["data"][:, (2,3)]
```

```python
y = iris["target"]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis", edgecolor="k")

# Adicionar rótulos
plt.xlabel("Comprimento da Pétala (cm)")
plt.ylabel("Largura da Pétala (cm)")
plt.title("Relação entre Comprimento e Largura da Pétala por Classe")

# Mostrar o gráfico
plt.show()
```

```python
iris_setosa = (y==0)
iris_versicolor = (y==1)

setosa_versicolor = iris_setosa | iris_versicolor

X = X[setosa_versicolor]
y = y[setosa_versicolor]

from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler

svc = LinearSVC(loss="hinge", C=10000, random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
svc.fit(X_scaled,y)

y_pred = svc.predict(X_scaled)
y_pred
```

```python

# Treinar o modelo LinearSVC
svc = LinearSVC(loss="hinge", C=5, random_state=42)
svc.fit(X_scaled, y)

# Calcular a inclinação (slope) e o viés (bias) da linha de decisão
w = -svc.coef_[0, 0] / svc.coef_[0, 1]  # Inclinação
b = -svc.intercept_[0] / svc.coef_[0, 1]  # Viés

# Transformar a linha de decisão de volta para a escala original
line = scaler.inverse_transform([[-10, -10 * w + b], [10, 10 * w + b]])

# Plotar os dados na escala original
plt.figure(figsize=(8, 6))
plt.plot(X[y == 0, 0], X[y == 0, 1], "yo", label="Iris setosa")
plt.plot(X[y == 1, 0], X[y == 1, 1], "bs", label="Iris versicolor")

# Plotar a linha de decisão
plt.plot(line[:, 0], line[:, 1], "k-", linewidth=2, label="LinearSVC Decision Boundary")

# Configurações do gráfico
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper center", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.show()
```
